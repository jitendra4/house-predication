```python
# find int and floating value
list_num_col=df.select_dtypes(include=['int64','float64']).columns
list_num_col
```

```python
# find object value 
list_obj_col=df.select_dtypes(include=['object']).columns
list_obj_col
```

```python
# find all missing value is it null
df.columns[df.isnull().any()] 
```

```python
def fillna_all(df):
    for i in list_obj_col:
        df[i].fillna(value=df[i].mode()[0],inplace=True)
    for i in list_num_col:
        df[i].fillna(value=df[i].mean(),inplace=True)
```

```python
fillna_all(df)
```
